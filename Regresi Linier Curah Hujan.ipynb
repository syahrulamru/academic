```python
import pandas as pd
import numpy as np
```

```python
# Masuk data & pengecekan awal
df = pd.read_csv('/content/Data_3000000009_2021.csv')
df
```

```python
# Menentukan kolom yang digunakan #
df= df[['temp','pressure','rh','solrad']]
df
```

```python
Feats= ['pressure','rh','solrad']

X= df[Feats]
y= df[['temp']]
```

```python
# Training dataset

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.4,random_state=42)
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
```

```python
# Fitting
MLR= LinearRegression()

MLR.fit(X_train, y_train)
```

```python
ent_press= float(input("Insert Pressure:"))
ent_rh= float(input("Insert Moisture:"))
ent_solrad= float(input("Insert Solar Radiation:"))

CatchtheRainbow= MLR.predict([[ent_press, ent_rh, ent_solrad]])
print("Predicted Temperature is {W}".format(W= CatchtheRainbow))
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

Rsq= MLR.score(X, y)
intercept= MLR.intercept_
coef= MLR.coef_

print("intercept={i}".format(i=intercept))
print("coefficient={c}".format(c=coef))
print("R squared={Rsq}".format(Rsq=Rsq * 100))
koef= list(zip([Feats], coef))
print("fitur & coef={k}".format(k=koef))

y_pred= MLR.predict(X)
MAE= mean_absolute_error
MSE= mean_squared_error

N= len(df)
RSEd= (y - y_pred)**2
RSE= np.sqrt(np.sum(RSEd)/N)
RMSE= MSE(y, y_pred, squared=False)
ymean= np.mean(y)
error= RSE/ymean
r_error= RSE/error
MAE_error= MAE(y, y_pred)
MSE_error= MSE(y, y_pred)
print("RSE = {R}".format(R=RSE))
print("RMSE = {RM}".format(RM=RMSE))
print("error= {e}".format(e=error))
print("error percentage= {e}".format(e=r_error))
print("MAE = {mae}".format(mae=MAE_error))
print("MSE = {mse}".format(mse=MSE_error))
```
